```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state =1, max_depth = 4)
model.fit(normalized_x_trained, X_train)
```

```python
#Get feature importance variables

features = df.columns
importances = model.feature_importances_
indices = np.argsort(importances)[-4:]
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color = "b", align = "center")
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel("Realtive importance")
plt.show
```

```python
y_predict = new_df.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print("MAE : ", mean_absolute_error(y_test, y_predict))
print("MSE : ", mean_squared_error(y_test, y_predict))
print("R2 score : ", r2_score(y_test, y_predict))
```

```python
train_score = []
test_score = []

random_state = 10

for i in range(random_state):
    X_train, X_test, y_train, y_test = train_test_split(X_val, y_val, test_size = 0.30, random_state = i)

    linear_model = LinearRegression()
    linear_model.fit(X_train, y_train)
    
    y_train_predict_mimax = linear_model.predict(X_train)
    y_test_predict_minmax = linear_model.predict(X_test)
    
    train_score = np.append(train_score, r2_score(y_train, y_train_predict_mimax))
    test_score = np.append(test_score, r2_score(y_test, y_test_predict_minmax))

print(X_train.shape)
print(y_train.shape)
```

```python
train_score
```

```python
test_score
```

```python
print('R2 train: %.2f +/- %.2f' % (np.mean(train_score),np.std(train_score)))
```

```python
print('R2 test: %.2f +/- %.2f' % (np.mean(test_score),np.std(test_score)))
```
